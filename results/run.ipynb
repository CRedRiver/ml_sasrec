```python
from evaluate.evaluate import evaluate_model
import numpy as np

import torch
from sasrec.preprocess import preprocess
from sasrec.sasrecnet import SASRecNet
from sasrec.sasrecdataset import SASRecDataset
from torch.utils.data import DataLoader
```

```python
DATA_PATH = r"D:\HUST\2025.2\ML\Project\data\ratings.csv"
GENRE_PATH = r"D:\HUST\2025.2\ML\Project\data\movies.csv"
seqs, num_items, item_map, movie_to_genre, genres_num = preprocess(DATA_PATH, GENRE_PATH)
dataset = SASRecDataset(seqs, genres_num, movie_to_genre, max_len=50)
dataloader = DataLoader(dataset, batch_size=64, shuffle=False)
```

```python
models = ["_67"]
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
for model in models:
    sasrec_model = SASRecNet(num_items=num_items,
                  max_len=50,
                  hidden_size=128,
                  num_heads=2,
                  dropout_rate=0.3,
                  num_genres=genres_num).to(device)
    path = f"../sasrec_BEST_weights{model}.pth"
    state_dict = torch.load(path, weights_only=True)

    sasrec_model.load_state_dict(state_dict)
    print(f"Model{model}")
    evaluate_model(sasrec_model, dataloader=dataloader,device=device,k=20,print_res=True)
```
